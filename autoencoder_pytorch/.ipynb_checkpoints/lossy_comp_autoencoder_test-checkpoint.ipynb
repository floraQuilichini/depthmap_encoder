```python
class ImageDataset(Dataset):

    def __init__(self, root_dir, transform=None):
        """
        Args:
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        self.root_dir = root_dir
        self.transform = transform

    def __len__(self):
        path, dirs, files = next(os.walk(self.root_dir))
        file_count = len(files)
        return file_count

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        path, dirs, files = next(os.walk(self.root_dir))
        img_name = os.path.join(self.root_dir, files[idx])
        image = io.imread(img_name)

        if self.transform:
            image = self.transform(image)

        return image


class RandomCrop(object):
    """Crop randomly the image

    Args:
        output_size (tuple or int): Desired output size. If int, square crop
            is made.
    """

    def __init__(self, output_size):
        assert isinstance(output_size, (int, tuple))
        if isinstance(output_size, int):
            self.output_size = (output_size, output_size)
        else:
            assert len(output_size) == 2
            self.output_size = output_size

    def __call__(self, image):

        h, w = image.shape[:2]
        new_h, new_w = self.output_size

        top = np.random.randint(0, h - new_h)
        left = np.random.randint(0, w - new_w)

        image = image[top: top + new_h,
                      left: left + new_w]


        return image
    

class ToTensor(object):
    """Convert ndarrays in images to Tensors."""

    def __call__(self, images):

        # swap color axis because
        # numpy image: H x W x C
        # torch image: C X H X W
        if (images.ndim ==3):
            images = images.transpose((2, 0, 1))
            return torch.from_numpy(images)
        else:
            images = np.expand_dims(images, axis=0)
            images = np.concatenate((images, images, images), axis=0)
            return torch.from_numpy(images.astype(np.int16))


        
class TorchStandardScaler(object):
    def fit(self, x):
        self.mean = ((x.to(torch.float)).mean((0,2,3), keepdim=True)).to(torch.short)
        self.std = ((x.to(torch.float)).std((0,2,3), unbiased=False, keepdim=True)).to(torch.short)
        print("mean : ", self.mean)
        print("std : ", self.std)
        
    def transform(self, x):
        if (x.is_cuda):
            x = x - (self.mean).to('cuda')
            x =x/((self.std).to('cuda') + 1e-7)
        else:
            x = x - (self.mean).to('cpu')
            x = x/(self.std + 1e-7).to('cpu')
        return x
    
    def inverse_transform(self, x):
        if (x.is_cuda):
            x = x*((self.std).to('cuda') + 1e-7)
            x = x+(self.mean).to('cuda')
        else:
            x = x*((self.std).to('cpu') + 1e-7)
            x = x+(self.mean).to('cpu')
        return x

        
        
class RandomContrastStreching2(object):
    """Equalize the histogram of the given image randomly with a given probability.
    If the image is torch Tensor, it is expected
    to have [..., 1 or 3, H, W] shape, where ... means an arbitrary number of leading dimensions.

    Args:
        p (float): probability of the image being equalized. Default value is 0.5
    """

    def __init__(self, p=0.5, a=0, b=2**(16)-1, percentile1=0, percentile2=100, bins=2**(16)):
        super().__init__()
        self.p = p
        self.a = a
        self.b =b
        self.percentile1 = percentile1
        self.percentile2 = percentile2
        self.bins = bins

    def __call__(self, img):
        """
        Args:
            img (Tensor): Image to be equalized.

        Returns:
            Tensor: Randomly equalized image.
        """

        if torch.rand(1).item() < self.p:
            img = img.to(torch.float)
            h = torch.histc(img, self.bins, min=torch.min(img), max=torch.max(img))
            nb_val = torch.sum(h)
            factor = 100.0/nb_val
            percentages = h*factor
            cum_sum = torch.cumsum(percentages, dim=0)
            
            # get lower value corresponding to the smaller percentile
            k1 = (cum_sum>self.percentile1).nonzero(as_tuple=True)[0][0]
            c = torch.min(img) + k1*(torch.max(img) - torch.min(img))/self.bins
            """
            for k in range(len(cum_sum)):
                if cum_sum[k] > self.percentile1:
                    c = torch.min(img) + k*(torch.max(img) - torch.min(img))/self.bins 
                    break
            """
            # get higher value corresponding to the bigger percentile
            k2 = (cum_sum>=self.percentile2).nonzero(as_tuple=True)[0][0]
            d = torch.min(img) + k2*(torch.max(img) - torch.min(img))/self.bins
            """
            for k in range(len(cum_sum)):
                if cum_sum[k] > self.percentile2:
                    d = torch.min(img) + k*(torch.max(img) - torch.min(img))/self.bins 
                    break
                d = torch.max(img)
            """
            img_streched = (img - c)*((self.b - self.a)/(d - c)) + self.a
            img_streched = torch.clamp(img_streched, min=self.a, max=self.b)
            #return exposure.equalize_adapthist(img.astype(float), kernel_size=None, clip_limit=0.02, nbins=10000)
            return img_streched.to(torch.short)
        return img


    def __repr__(self):
        return self.__class__.__name__ + '(p={})'.format(self.p)
    


    
class RandomApply(object):
    """Randomly apply a transformation between a list of transformations given as input
    """

    def __init__(self, list_of_transfomations):
        super().__init__()
        self.list = list_of_transfomations

    def __call__(self, img):
        """
        Args:
            img (Tensor): Image to be equalized.

        Returns:
            Tensor: Randomly equalized image.
        """

        list_length = len(self.list)
        init = 1.0/list_length
        prob = torch.rand(1).item()
        bound_val = torch.linspace(init, 1.0, steps=list_length)
        
        k = (bound_val>=prob).nonzero(as_tuple=True)[0][0]
        
        """
        val = step
        ind = int(list_length) - 1
        for k in range(list_length):
            if (val>prob):
                ind = k
                break
            val = val + step
        
        print(k)
        """
        return self.list[k](img)
                


    def __repr__(self):
        return self.__class__.__name__  
    
       

class Normalize(object):
    """Normalize ndarrays"""

    def __call__(self, images):

        # numpy image: H x W x C
        min_val = images.min(axis=(0, 1))
        max_val = images.max(axis=(0, 1))
        
        normalized_im = (images.astype(float) - min_val.astype(float))/(max_val.astype(float) - min_val.astype(float))
        #normalized_im  = 2*(normalized_im - 0.5)
        
        return normalized_im
    
        
        
def show_image_batch(images_batch):
    """Show image for a batch of samples."""
    if (images_batch.size()[1] == 3):
        print("max value", torch.max(images_batch))
        print("min value", torch.min(images_batch))
        images_batch_normed = images_batch.double()/(torch.max(images_batch)).double()
        grid = utils.make_grid(images_batch_normed)
        plt.imshow(grid.numpy().transpose((1, 2, 0)))
    else:
        print("max value", torch.max(images_batch))
        print("min value", torch.min(images_batch))
        images_batch_normed = images_batch.double()/(torch.max(images_batch)).double()
        grid = utils.make_grid(images_batch_normed)
        plt.imshow(np.squeeze(grid.numpy().transpose(1,2,0)), cmap='gray')
    plt.title('Batch from dataloader')
    
    
    
    
# Load training dataset

#mirror_padding_transform = transforms.Compose([transforms.ToPILImage(), transforms.Pad(padding=12, padding_mode='reflect'), transforms.ToTensor()])
"""
transformed_dataset = ImageDataset('D:/autoencoder_data/depthmaps/training/dilated/', #'./data/flickr/',
                                           transform=transforms.Compose([
                                            RandomCrop(128), ToTensor()])
                                  )

"""
transformed_dataset = ImageDataset('./data/flickr/', #'D:/autoencoder_data/depthmaps/training/mixed/', 
                                           transform=transforms.Compose([
                                            ToTensor(),
                                            RandomApply([transforms.Resize(((128, 128))),
                                                          transforms.RandomResizedCrop((128, 128), scale=(0.08, 1.0), ratio=(0.75, 2.3333333333333333)),
                                                          #transforms.Compose([transforms.RandomAffine(30, translate=(0.1, 0.1), scale=(0.7, 1.3), shear=None), transforms.RandomCrop(128)]), 
                                                          #transforms.Compose([transforms.RandomAffine(30, translate=(0.1, 0.1), scale=(0.5, 1.5), shear=None), transforms.Resize((128, 128))]),
                                                          transforms.Compose([transforms.RandomErasing(p=0.5, scale=(0.02, 0.15), ratio=(0.3, 3.3), value=0), transforms.Resize((128, 128))]) 
                                                          #transforms.Compose([RandomContrastStreching2(p=0.8, a=0, b=2**(16)-1, percentile1=20, percentile2=80, bins=2000) , transforms.RandomResizedCrop((128, 128), scale=(0.08, 1.0), ratio=(0.75, 2.3333333333333333))]) 
                                                        ])])
                                  )
print(transformed_dataset[1].size())

#get mean and std from training dataset
dataloader_init = DataLoader(transformed_dataset, batch_size=840, shuffle=True, num_workers=0)
for i_batch, batch_images in enumerate(dataloader_init):
    foo = TorchStandardScaler()
    foo.fit(batch_images)


# dataloader
dataloader = DataLoader(transformed_dataset, batch_size=64, shuffle=True, num_workers=0)

for i_batch, batch_images in enumerate(dataloader):
    print(i_batch, batch_images.size())
    #print(len(batch_images.size()))

    # observe 4th batch and stop.
    if i_batch == 2:
        plt.figure()
        show_image_batch(batch_images)
        plt.axis('off')
        plt.ioff()
        plt.show()
        break

"""
transform = transforms.Compose([
    transforms.Pad(12, padding_mode='reflect'),
    transforms.ToTensor()])

trainset = torchvision.datasets.Flickr8k(root='./data/flickr', ann_file = './data/flickr/file.csv', transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=True, num_workers=2)
"""
```

```python
#Define the Convolutional Autoencoder with integrated classifer
    #taille de l'image d'entre : 256*256
class LossyCompAutoencoder(nn.Module):
    def __init__(self):
        super(LossyCompAutoencoder, self).__init__()
       
        #Encoder
            # input block
        self.conv1 = nn.Conv2d(3, 64, 5, stride=2, padding=0)  
        self.conv2 = nn.Conv2d(64, 128, 5, stride=2, padding=0)
            # residual block 
        self.resConv = nn.Conv2d(128, 128, 3, stride=1, padding=1)
            # output block
        self.conv3 = nn.Conv2d(128, 96, 5, stride=2, padding=0)
        self.quantization = MyQuantization.apply
        #self.gaussian_distribution = GaussianDistribution.apply
        

       
        #Decoder
            # subpixel 1
        self.subpix1 = nn.Conv2d(96, 512, 3, stride=1, padding=1)
            # subpixel 2
        self.subpix2 = nn.Conv2d(128, 256, 3, stride=1, padding=1)
            # subpixel 3
        self.subpix3 = nn.Conv2d(256//4, 12, 3, stride=1, padding=1)
            #residual block
        self.deconv1 = nn.Conv2d(512//4, 128, 3, stride=1, padding=1)
        self.deconv2 = nn.Conv2d(128, 128, 3, stride=1, padding=1)
        self.clip = MyClipping.apply
        
        #Bit-rate      
        self.var = nn.Parameter(torch.Tensor(6, 96))
        self.phi = nn.Parameter(torch.Tensor(6, 96))
        self.var.data.uniform_(0, 1)
        self.phi.data.uniform_(0, 1)
        
        """
        self.gsm_pi = torch.nn.Parameter(torch.randn(6, 96))
        self.gsm_sigma = torch.nn.Parameter(torch.randn(6, 96))
        """
    
    
    
    def forward(self, x, foo, mask= 1, return_xq=False):
        #encoder
            # normalization
        #x, radius, mean_channels = normalize_input(x)
        x = foo.transform(x)
        #print("radius : ", radius)
        #print("mean_channel :", mean_channels)
            # mirror padding
        x = mirror_padding(x, 14)
        """
        with torch.no_grad():
            x_copy = x.cpu()
            show_image_batch(x_copy)
        """
            # input blocks
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x_c1 = x.clone()
            # residual block 1
        x = F.relu(self.resConv(x))
        x = self.resConv(x)
        x += x_c1
        x_c2 = x.clone()
            # residual block 2
        x = F.relu(self.resConv(x))
        x = self.resConv(x)
        x += x_c2
        x_c3 = x.clone()
            # residual block 3
        x = F.relu(self.resConv(x))
        x = self.resConv(x)
        x += x_c3
            # output block
        x = self.conv3(x)
            # quantization
        x = self.quantization(x)
            # add mask for incremental training
        x = x*mask
        x_quantized = x
        
        
        """
        w = x_quantized.size()[2]
        h = x_quantized.size()[3]
        gsm = 0.0
        for i in range(6)
            mi = torch.flatten(x_quantized),torch.diagonal(self.gsm_sigma[s, :].repeat(w*h, 1).squeeze(0))
            gsm += 
        """
        

        #decoder
            # subpixel 1
        x = self.subpix1(x)
        x = periodic_shuffling(x, 512//4)
        x_c4 = x.clone()
            # residual block 1
        x = F.relu(self.deconv1(x))
        x = self.deconv2(x)
        x += x_c4
        x_c5 = x.clone()
               # residual block 2
        x = F.relu(self.deconv2(x))
        x = self.deconv2(x)
        x += x_c5
        x_c6 = x.clone()
               # residual block 3
        x = F.relu(self.deconv2(x))
        x = self.deconv2(x)
        x += x_c6
                # subpixel 2
        x = self.subpix2(x)
        x = F.relu(periodic_shuffling(x, 256//4))
                # subpixel 3
        x = self.subpix3(x)
        x = periodic_shuffling(x, 12//4)
                # denormalization
        #x = denormalize_output(x, radius, mean_channels)
        x = foo.inverse_transform(x)
                # clipping
        x = self.clip(x)

        
        if return_xq:
            return x, x_quantized
        else:
            return x
      
```

```python
def entropy_rate(x_quantized, phi, var):
    u = torch.arange(-0.5, 0.5+0.01, 0.01).cuda()        
    gsm_sum = torch.zeros(len(u)).cuda()
    for i in range(len(u)):
        x = x_quantized + u[i]
        gsm_sum_i = sum_gsm(x, var, phi, 6)
        gsm_sum[i] = gsm_sum_i

    entropy = torch.trapz(gsm_sum, u)
    
    return entropy


# Load incremental model
incremental_model = LossyCompAutoencoder()
incremental_model.load_state_dict(torch.load('./model_parameters/lossy_comp_params_with_rate_beta2_incremental.pth'))
incremental_model.eval()
incremental_model.to(device)

# train again the model starting form incremental-learned weights
    #define optimizer
optimizer = torch.optim.Adam(incremental_model.parameters(), lr=0.0001)

# define loss function
distortion = nn.MSELoss().cuda()


# define beta
beta = 2.0

#Epochs
n_epochs = 600


# Training the network
for epoch in range(1, n_epochs+1):
    running_loss = 0.0
           
    #Training
    for i_batch, data in enumerate(dataloader):
        batch_images = data.to(device).float()
        [decoded_images, x_quantized] = incremental_model(batch_images, False, True)
        optimizer.zero_grad()
        entropy = entropy_rate(x_quantized, incremental_model.phi, incremental_model.var)
        print("entropy : ", entropy)
        dist = distortion(decoded_images, batch_images)
        print("distortion : ", dist)
        loss = beta * dist + entropy
        loss.backward()
        #print("conv1.weights grad: ", params[0].grad)
        #print(model.conv1.bias.grad)
        #print(model.conv1.weight.grad)
        optimizer.step()
        running_loss += loss.item()

    running_loss = running_loss/len(dataloader)
    print('running loss : {:.06f}'.format(running_loss), )
    
```
