```python
import numpy as np
import torch
import torch.nn as nn
import torch.autograd as ag
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data.sampler import SubsetRandomSampler
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import os
from scipy import signal
from sklearn.metrics import mean_squared_error
from torch.autograd import Variable
import time

torch.set_printoptions(precision=6)


def readDepthMap(depthMap_filename, depthmap_stack, batch_size, k):
    depth_file = open(depthmap_filename, "rb")
    height = int.from_bytes(depth_file.read(4), byteorder='little')
    width = int.from_bytes(depth_file.read(4), byteorder='little')
    depthmap = np.frombuffer(depth_file.read(), dtype=np.float32)
    depthmap.shape = (width, height)
    depthmap = np.flip(depthmap,axis=0)
    if isinstance(depthmap_stack, list):
        depthmap_stack[k // batch_size][k%batch_size, :, :,:] = torch.from_numpy(np.ascontiguousarray(depthmap))
    else:
        depthmap_stack[k :,:,:] = torch.from_numpy(np.ascontiguousarray(depthmap))
    
    #plt.imshow(training_depthmap, cmap='gray', vmin=0, vmax=1)
    depth_file.close()
    
def tile(a, dim, n_tile):
    init_dim = a.size(dim)
    repeat_idx = [1] * a.dim()
    repeat_idx[dim] = n_tile
    a = a.repeat(*(repeat_idx))
    order_index = torch.LongTensor(torch.cat([init_dim * torch.arange(n_tile) + i for i in range(init_dim)])).to(device='cuda:0')
    return torch.index_select(a, dim, order_index)


def accuracy(v1, v2):
    v1 = torch.squeeze(v1)
    v2 = torch.squeeze(v2)
    if (len(v1.size()) != 1 | len(v2.size()) !=1):
        print("error, v1 and v2 should be vectors")
        return -1
    else:
        if len(v1) != len(v2):
            print("error, v1 and v2 should have the same number of elements")
            return -1
        else:
            accuracy = 0
            for k in range(len(v1)):
                if v1[k] == v2[k]:
                    accuracy += 1
            return accuracy/len(v1)
        


# change dynamic range of image
def change_grey_dynamic(batch_images, max_dynamic):
    batch_size = len(batch_images)
    height = batch_images.size()[2]
    width = batch_images.size()[3]
    max_value = torch.max(batch_images)
    min_value = torch.min(batch_images)
    #inverse the gray levels and put values between [-max_dynamic/2, max_dynamic/2]
    #dynamic_changed = torch.floor((1-(batch_images - min_value)/(max_value - min_value))*max_dynamic) - (max_dynamic+1)//2
    dynamic_changed = (1-batch_images)*max_dynamic #- (max_dynamic+1)//2
    #dynamic_changed[(dynamic_changed == - (max_dynamic+1)//2).nonzero()] = 0;
    return dynamic_changed


def image_corr(batch_images, outputs):
    batch_size = len(batch_images)
    #print(batch_size)
    height = batch_images.size()[2]
    width = batch_images.size()[3]
    lags_mat = np.zeros((batch_size,2))
    maxCorr = np.zeros((1, batch_size))
    for k in range(batch_size):
        #cor = F.conv2d(batch_images[k, :,:,:].view(1, 1, height, width), torch.flip(torch.unsqueeze(outputs[k, :,:,:], 0), [2 , 3]), stride = [1, 1], padding = [height//2, width//2])
        cor = F.conv2d(batch_images[k, :,:,:].view(1, 1, height, width), torch.unsqueeze(outputs[k, :,:,:], 0), stride = [1, 1], padding = [height//2, width//2])
        max_corr = torch.max(cor)
        maxCorr[0, k] = max_corr
        max_lag = np.unravel_index((torch.argmax(torch.squeeze(cor), axis=None)).cpu(), torch.squeeze(cor).shape)
        #print(max_lag)
        lag_i = torch.argmax(torch.max(torch.squeeze(cor), 1)[0])
        #print("line : ")
        #print(lag_i)
        lag_j = (torch.argmax(torch.squeeze(cor), axis=1))[lag_i]
        #print("col : " )
        #print(lag_j)
        lags_mat[k, :] = max_lag
        #if max_lag[0] != 25 | max_lag[1] != 25 :
            #print(max_lag)
            #print(lag_i, lag_j)
            
        #print(max_lag)
    return lags_mat, maxCorr


def getMostSimilarBottleneckVector(query_vectors, database_vectors, metric_type):
    if metric_type == "L2":
        """ # vectorized implementation
        
        #print(database_vectors.repeat([len(query_vectors)] + [1 for x in database_vectors.shape[1:]]).size())
        #print(((database_vectors.repeat([len(query_vectors)] + [1 for x in database_vectors.shape[1:]]) - tile(query_vectors, 0, len(database_vectors)))**2).size())
        #print(((database_vectors.repeat([len(query_vectors)] + [1 for x in database_vectors.shape[1:]]) - tile(query_vectors, 0, len(database_vectors)))**2).view((len(database_vectors)*len(query_vectors), -1)).size())
        res = torch.sqrt(torch.sum(((database_vectors.repeat([len(query_vectors)] + [1 for x in database_vectors.shape[1:]]) - tile(query_vectors, 0, len(database_vectors)))**2).view((len(database_vectors)*len(query_vectors), -1)), 1))
        #print(res.size())
        #print(torch.argmin(res.view((len(query_vectors),len(database_vectors))), axis = 1).size())
        return torch.argmin(res.view((len(query_vectors),len(database_vectors))), axis = 1)
        """
        indices = torch.zeros(1, len(query_vectors))
        for k in range(len(query_vectors)):
            res = query_vectors[k, ...].view(1, -1) - database_vectors
            dists = torch.sqrt(torch.sum(res**2, 1))
            indices[0, k] = torch.argmin(dists)
        return indices

    else:
        print("error, this metric doesn't exists or is not implemented yet")
        return -1
    
def getMostSimilarDepthMap(depthMaps_query, depthMaps_ref, metric_type):
    nb_query = len(depthMaps_query)
    nb_ref = len(depthMaps_ref)
    height = depthMaps_query.size()[2]
    width = depthMaps_query.size()[3]
    depthMaps_query_stacked = tile(depthMaps_query, 0, nb_ref)
    depthMaps_ref_stacked = depthMaps_ref.repeat([nb_query] + [1 for x in depthMaps_ref.shape[1:]])
    if metric_type== "L2":
        s = torch.sqrt(torch.sum(torch.squeeze((depthMaps_query_stacked - depthMaps_ref_stacked)**2), (2, 1)))
        s = s.view(nb_query, nb_ref)
        most_sim_ind = torch.argmin(s, dim=1)
        #print(most_sim_ind)
        return depthMaps_ref[most_sim_ind, :, :, :]
    else : 
        if metric_type== "corr":
            depthMaps_query_stacked_new_dyn = change_grey_dynamic(depthMaps_query_stacked, 255)
            depthMaps_ref_stacked_new_dyn = change_grey_dynamic(depthMaps_ref_stacked, 255)
            lags_mat, maxCorr = image_corr(depthMaps_ref_stacked_new_dyn, depthMaps_query_stacked_new_dyn)
            lags_mat = lags_mat.astype(int)
            maxCorr = maxCorr.reshape((nb_query, nb_ref))
            #print(maxCorr)
            corr_ind = np.argmax(maxCorr, axis=1)
            #print(corr_ind)
            #print(np.size(maxCorr, 1))
            #return depthMaps_ref[corr_ind, :, :, :]
        
            # corr indices
            s = torch.zeros(nb_ref*nb_query, 1)
            for k in range(nb_ref*nb_query):
                depthMaps_ref_i = [max(0, lags_mat[k, 0]- height//2), min(height, height//2 + lags_mat[k, 0] +1)]
                depthMaps_ref_j = [max(0, lags_mat[k, 1]- width//2), min(width, width//2 + lags_mat[k, 1] +1)]
                depthMaps_query_i = [max(0, height//2 -lags_mat[k, 0]), min(height, height + height//2-lags_mat[k, 0])]
                depthMaps_query_j = [max(0, width//2 -lags_mat[k, 1]), min(width, width + width//2-lags_mat[k, 1])]
                s[k, 0] = torch.sqrt(torch.sum(torch.squeeze((depthMaps_query_stacked[k,:,depthMaps_query_i[0]:depthMaps_query_i[1],depthMaps_query_j[0]:depthMaps_query_j[1]] - depthMaps_ref_stacked[k,:,depthMaps_ref_i[0]:depthMaps_ref_i[1], depthMaps_ref_j[0]:depthMaps_ref_j[1]])**2)))
            s = s.view(nb_query, nb_ref)
            #print(s)
            #print(s[0, 156])
            #print(s[0, torch.argmin(s, dim=1)])
            most_sim_ind = torch.argmin(s, dim=1)
            #print(most_sim_ind)
            return depthMaps_ref[most_sim_ind, :, :, :]
        else:
            return -1
```

```python
#Define the Convolutional Autoencoder with integrated classifer
    #taille de l'image d'entre : 256*256
class ConvAutoencoderWithClassifier(nn.Module):
    def __init__(self):
        super(ConvAutoencoderWithClassifier, self).__init__()
       
        #Encoder
        self.fc1 = nn.Linear(128**2, 50**2)
        self.fc2 = nn.Linear(50**2, 30**2)
        self.fc3 = nn.Linear(30**2, 10**2)
        #self.fc4 = nn.Linear(10**2, 50)
        #self.fc5 = nn.Linear(50, 11)
        self.pool = nn.MaxPool2d(2, 2)
       
        #Decoder
        #self.t_fc1 = nn.Linear(11, 50)
        #self.t_fc2 = nn.Linear(50, 10**2)
        self.t_fc3 = nn.Linear(10**2, 30**2)
        self.t_fc4 = nn.Linear(30**2, 50**2)
        self.t_fc5 = nn.Linear(50**2, 256**2)
        
        #Classifier
        self.fc4 = nn.Linear(10**2, 50)
        self.fc5 = nn.Linear(50, 11)
        


    def forward(self, x, return_comp=False):
        #encoder
        # add hidden layers with relu activation function
        x = self.pool(x).view(-1, 1*128*128)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        #x = F.relu(self.fc4(x))
        #x = F.relu(self.fc5(x))
        x_comp = x  # compressed representation

        
        #decoder
        # add transpose conv layers, with relu activation function
        
        #x = F.relu(self.t_fc1(x))
        #x = F.relu(self.t_fc2(x))
        x = F.relu(self.t_fc3(x))
        x = F.relu(self.t_fc4(x))
        x = F.relu(self.t_fc5(x))
        x = x.view(-1, 1, 256, 256)
        
        # classifier
        x_class = F.relu(self.fc4(x_comp))
        x_class = F.relu(self.fc5(x_class))
   
        if return_comp:
            return x, x_class, x_comp
        else:
            return x, x_comp
        


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # images of size 32*1*256*256 (size batch: 32 , nb channels : 1 , h : 256, w : 256)
        self.conv1 = nn.Conv2d(1, 6, 5) #output : 32*6*252*252
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)  #output : 32*16*122*122
        self.fc1 = nn.Linear(16 * 61 * 61, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 11)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))  #output : 32*6*126*126
        x = self.pool(F.relu(self.conv2(x)))  #output : 32*16*61*61
        x = x.view(-1, 16 * 61 * 61)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x


#model = Net()
        

# instanciate the model
model = ConvAutoencoderWithClassifier()
print(model)
params = list(model.parameters())
print(len(params))
print(params)
   

```

```python
def get_device():
    if torch.cuda.is_available():
        device = 'cuda:0'
    else:
        device = 'cpu'
    return device

device = get_device()
print(device)
model.to(device) # transfert du model au gpu
```

```python
"""
#criterion = nn.BCELoss()
#criterion = nn.MSELoss()
criterion_1 = nn.CrossEntropyLoss().cuda()
criterion_2 = nn.MSELoss().cuda()

#Optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
"""

criterion_1 = nn.CrossEntropyLoss().cuda()
criterion_2 = nn.MSELoss().cuda()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9) # stochastic gradient descent for weights updates


#Epochs
n_epochs = 100
lfactor = 0.1

# losses to print
classif_losses = torch.zeros(1, len(training_dataset)*n_epochs)
reconstruction_losses = torch.zeros(1, len(training_dataset)*n_epochs)
total_losses = torch.zeros(1, len(training_dataset)*n_epochs)
train_losses = torch.zeros(1, n_epochs)

# Training the network
for epoch in range(1, n_epochs+1):
    # monitor training loss
    train_loss = 0.0
    k = 0
    #if epoch==30:
        #lfactor = 0.9

    #Training
    for data in training_dataset:
        print("training set n", k )
        batch_images = data.to(device)
        labels = training_labels[k].type(torch.LongTensor).to(device)
        #print(labels.size())
        [decoded_images, classification_output, training_latent_features] = model(batch_images, True)
        #classification_output = model(batch_images)
        print(classification_output.size())
        #print(decoded_images.size())
        optimizer.zero_grad()
        loss = lfactor*criterion_1(classification_output, labels) + (1-lfactor)*criterion_2(decoded_images, batch_images)
        #loss = criterion(classification_output, labels)
        #print("erreur de reconstruction : " , criterion_2(decoded_images, batch_images))
        #print("erreur de classification : ", criterion_1(classification_output, labels))
        #print(loss)
        with torch.no_grad():
            classif_losses[0, (epoch-1)*n_epochs+k] = criterion_1(classification_output, labels)
            reconstruction_losses[0, (epoch-1)*n_epochs+k] = criterion_2(decoded_images, batch_images)
            total_losses[0, (epoch-1)*n_epochs+k] = loss
        
        loss.backward()
        #print("conv1.weights grad: ", params[0].grad)
        #print(model.conv1.bias.grad)
        #print(model.conv1.weight.grad)
        optimizer.step()
        train_loss += loss.item()*batch_images.size(0)
        k = k+1
          
    train_loss = train_loss/len(training_dataset)
    with torch.no_grad():
        train_losses[0, epoch-1] = train_loss
    print('Epoch: {} \tTraining Loss: {:.6f}'.format(epoch, train_loss))
```
